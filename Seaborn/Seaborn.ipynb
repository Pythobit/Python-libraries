# Distribution Plots

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # kind='reg' stands for regression, and kind='kde'
```

```python
sns.pairplot(tips, hue='sex') #palette='coolwarm'
```

```python
sns.rugplot(tips['total_bill'])
```

- Study KDE(Kernel Density Estimation), and NE(Normal Estimation)
# Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin') # you can use catplot also, it gives it the same operation
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
grid = sns.FacetGrid(data=tips, row='smoker', col='time')
grid.map(sns.distplot, 'total_bill')
```
