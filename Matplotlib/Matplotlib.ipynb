```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional way

plt.plot(x, y)
plt.xlabel('This is x-axis title')
plt.ylabel('This is y-axis title')
plt.title('This is Main title of the plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(x, y, 'b')
```

```python
# OO : Object Oreiented

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

# Figure size & DPI

```python
fig = plt.figure(figsize=(6,3))

axe = fig.add_axes([0,0,1,1])
axe.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2 ,ncols=1, figsize=(6,3))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure()

axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
axe.plot(y,x)
```

```python
fig = plt.figure()

axe = fig.add_axes([0,0,1,1])
axe.plot(x,x**2, label='Square of x')
axe.plot(x,x**3, label='Cube of x')

axe.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()

axe = fig.add_axes([0, 0, 1, 1])
axe.plot(x, y, color='maroon', lw=2, alpha=0.6, ls='--', marker='1', markersize=20, 
         markerfacecolor='CadetBlue', markeredgewidth=3, markeredgecolor='blue')
```

```python
fig = plt.figure()

axe = fig.add_axes([0, 0, 1, 1])
axe.plot(x, y, color='maroon', lw=2, ls='--')

axe.set_xlim([0,1])
axe.set_ylim([0,2])
```
